```python
def check(nums):
    total=sum(nums)
    Average=total/len(nums)
    Maxi=max(nums)
    Mini=min(nums)
    count=len(nums)
    even=[]
    odd=[]
    for  i in nums:
        if i%2==0:
            even.append(i)
        elif i%2!=0:
            odd.append(i)
    return total, Average, Maxi, Mini, count, even, odd
def main():
    nums=input("Enter the values: ").split()
    nums=[int(i) for i in nums]
    result=check(nums)
    print("\nResult")
    print(f"Sum: {result[0]}")
    print(f"Average: {result[1]}")
    print(f"Max: {result[2]}")
    print(f"Min: {result[3]}")
    print(f"Count: {result[4]}")
    print(f"Even: {result[5]}")
    print(f"Odd: {result[6]}")
main()
```
